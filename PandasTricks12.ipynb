# Trick 5 - Multiple conditions Query

```python
sex_male = 'male'
age_30 = 30
deck_ABC = ['A', 'B', 'C']

df.query('alone == 1 & alive == "no" & sex == @sex_male & age == @age_30 & deck == @deck_ABC')

# df[df['deck'].isin(deck_ABC)]
```

# Trick 6 - Label Encoding

```python
df['deck'].unique().sort_values()
```

```python
dict_deck = dict(zip(df['deck'].unique().sort_values(), range(8)))
dict_deck
```

```python
df['deck'] = df['deck'].map(dict_deck)
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['deck'] = le.fit_transform(df['deck'])
```

```python
le.classes_
```

#Trick 8 - Create Category Variables

```python
age_bins = [0, 13, 19, 61, np.inf]
age_labels = ['<= 13', 'Teen', 'Adult', 'Older']

df['ageGroup'] = pd.cut(df['age'], bins = age_bins, labels = age_labels)
```

```python
df['ageGroup'].value_counts().plot.pie()
```

# Trick 9 - transform Function with groupby()

```python
df['mean_fare_bySexPclass'] = df.groupby(['sex', 'pclass'])['fare'].transform('mean')
df
```

```python
df.isnull().sum()
```

```python
df["age"].fillna(df.groupby(['sex','pclass'])["age"].transform("median"), inplace=True)
```

```python
df.isnull().sum()
```

# Trick 10 - apply Function

```python
df['fareGroup'] = df['fare'].apply(lambda x: 'Low' if x >= 100 else
                                             'Mid' if x >= 10 else
                                             'High')
df.head()
```

# Trick 11 - Rename Columns after groupby()

```python
df_tmp = df.groupby('ageGroup').agg({'age': 'mean', 'fare': ['min', 'max']})
df_tmp
```

```python
df_tmp.columns
```

```python
df_tmp.columns = ['_'.join(col) for col in df_tmp.columns]
df_tmp
```
